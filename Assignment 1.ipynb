```python
# Question 1

import math

def isPrime(number):
        
    if number == 1 or number == 4:
        return bool(False)
    elif number == 2:
        return bool(True)
    
    if number > 2:
        prime = bool(True)
        numberSquared = math.ceil(math.sqrt(number)) + 1

        for divisor in range(2, numberSquared):
            if number % divisor == 0:
                prime = bool(False)
                break
        
        if prime == 1:
            return bool(True)
        else:
            return bool(False)
    else:
        print("Number must be positive")
        return bool(False)


circularPrimes = 0
print("Circular prime numbers: ", end = '')
for number in range(2, 10000):
    count = 0
    if isPrime(number):
        if number > 10:
            numberInString = str(number)
            for variation in range(1, len(numberInString)):
                numberInString = numberInString[1:] + numberInString[0:1]
                if isPrime(int(numberInString)):
                    count = count + 1
            if count == (len(numberInString) - 1):
                print(number, end = ' ')
                circularPrimes = circularPrimes + 1
        else:
            print(number, end = ' ')
            circularPrimes = circularPrimes + 1
            
print("\nTotal number of circular primes = ",circularPrimes)
```

```python
# Question 5

class MyCircularQueue():
# source: https://www.programiz.com/dsa/circular-queue
    def __init__(self, k):
        self.k = k
        self.queue = [None] * k
        self.head = self.tail = -1
        
    def isEmpty(self):
        if self.tail == -1:
            return bool(True)
        return bool(False)
    
    def isFull(self):
        if ((self.tail + 1) % self.k == self.head):
            return bool(True)
        return bool(False)

    # Insert an element into the circular queue
    def enqueue(self, data):

        if ((self.tail + 1) % self.k == self.head):
            print("The circular queue is full\n")

        elif (self.head == -1):
            self.head = 0
            self.tail = 0
            self.queue[self.tail] = data
        else:
            self.tail = (self.tail + 1) % self.k
            self.queue[self.tail] = data

    # Delete an element from the circular queue
    def dequeue(self):
        if (self.head == -1):
            print("The circular queue is empty\n")

        elif (self.head == self.tail):
            temp = self.queue[self.head]
            self.head = -1
            self.tail = -1
            return temp
        else:
            temp = self.queue[self.head]
            self.head = (self.head + 1) % self.k
            return temp
    
    def getFrontValue(self):
        if (self.head == -1):
            print("The circular queue is empty\n")
        else:
            return self.queue[self.tail]
            
    def printQueue(self):
        if(self.head == -1):
            print("No element in the circular queue")

        elif (self.tail >= self.head):
            for i in range(self.head, self.tail + 1):
                print(self.queue[i], end=" ")
            print()
        else:
            for i in range(self.head, self.k):
                print(self.queue[i], end=" ")
            for i in range(0, self.tail + 1):
                print(self.queue[i], end=" ")
            print()

sizeOfQueue = 10
circularQueue = MyCircularQueue(sizeOfQueue)

circularQueue.enqueue(3)
circularQueue.enqueue(8)
circularQueue.enqueue(9)
circularQueue.enqueue(1)
circularQueue.enqueue(2)
circularQueue.enqueue(7)
circularQueue.enqueue(4)
circularQueue.enqueue(6)
circularQueue.enqueue(5)
circularQueue.enqueue(10)

print("Queue:", end=' ')
circularQueue.printQueue()

if not circularQueue.isEmpty():
    
    currentElement = circularQueue.dequeue()
    minimum = currentElement
    lastValue = circularQueue.getFrontValue()

    while currentElement != lastValue:
        
        if currentElement < minimum:
            minimum = currentElement
        
        circularQueue.enqueue(currentElement)
        if not circularQueue.isEmpty():
            currentElement = circularQueue.dequeue()
            
    circularQueue.enqueue(currentElement)

    print("Minimum value:", minimum)
else:
    print("Queue is empty")
```

```python
# Question 6

def enqueue(queue, element):
    queue.append(element)

def dequeue(queue):
    if queue:
        return queue.pop(0)

def push(queue, element):
    enqueue(queue, element)

def pop(queue):
    if queue:
        helpingQueue = []
        for i in range(0, len(queue) - 1):
            enqueue(helpingQueue, dequeue(queue))
        poppedElement = dequeue(queue)
        for i in range(0, len(helpingQueue)):
            enqueue(queue, dequeue(helpingQueue))
        return poppedElement
    else:
        print("Stack is empty")
        
def printQueue(queue):
    print("Queue: ", end='')
    for element in queue:
        print(element, end=' ')
    print()
    
queue = []

for i in range(1,4):
    push(queue, i)
    printQueue(queue)
    
for i in range(1,5):
    print("Popped element:",pop(queue))
    printQueue(queue)
    
```
